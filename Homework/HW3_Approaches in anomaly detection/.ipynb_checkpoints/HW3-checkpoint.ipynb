```python
from sklearn.neighbors import KernelDensity
from sklearn.cross_validation import train_test_split
import numpy as np
import scipy.io
mat = scipy.io.loadmat('oc_514.mat')
# print dir(mat)
# print mat.keys()
# print mat['__version__']
# print mat['__header__']
# print mat['__globals__']
# print type(mat['x']),type(mat['z'])
# print mat['z']
# print mat['x']
train = mat['x']
# print train.shape
# print type(train)
# print dir(train)
# print dir(train[0,0])
# print train
# print len(train[0,0][0])#[-1])
xtrain = train[0,0][0]
# print (train[0,0][0])#[-1])
X_train, X_test = train_test_split(
    xtrain, test_size=0.33, random_state=42)
#Using Gaussian Kernel Density Estimation
pdf = KernelDensity(bandwidth=0.3)
pdf.fit(X_train)
res = []
# print pdf
for data in X_test:
#     print np.exp(pdf.score_samples(data))[0]
    res.append(np.exp(pdf.score_samples(data))[0])
print res
```
